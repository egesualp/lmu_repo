```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression

# Create a synthetic dataset with perfect separation
X, y = make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0, random_state=42)

# Fit logistic regression with theta
theta = np.array([1, -2])  # Just an example, you can change this
model_theta = LogisticRegression(solver='liblinear', random_state=42, C=1e5)
model_theta.fit(X, y)

# Fit logistic regression with 2*theta
model_2theta = LogisticRegression(solver='liblinear', random_state=42, C=1e5)
model_2theta.fit(X, y)

# Plot the data points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', marker='o', edgecolors='k', s=100, label='Data points')

# Plot the decision boundaries
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))

# Plot decision boundary for theta
Z_theta = model_theta.predict(np.c_[xx.ravel(), yy.ravel()])
Z_theta = Z_theta.reshape(xx.shape)
plt.contour(xx, yy, Z_theta, colors='r', linewidths=2, levels=[0.5], label='Decision Boundary (theta)')

# Plot decision boundary for 2*theta
Z_2theta = model_2theta.predict(np.c_[xx.ravel(), yy.ravel()])
Z_2theta = Z_2theta.reshape(xx.shape)
plt.contour(xx, yy, Z_2theta, colors='b', linewidths=2, levels=[0.5], label='Decision Boundary (2*theta)')

# Plot the vector theta and 2*theta
plt.arrow(0, 0, theta[0], theta[1], head_width=0.1, head_length=0.1, fc='g', ec='g', label='Theta')
plt.arrow(0, 0, 2 * theta[0], 2 * theta[1], head_width=0.1, head_length=0.1, fc='m', ec='m', label='2*Theta')

# Set plot limits
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())

# Add labels and legend
plt.title('Logistic Regression with Perfect Separation')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
