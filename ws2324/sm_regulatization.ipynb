```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Function to compute the Least Squares criterion
def least_squares_criterion(B, X, y):
    residuals = y - np.dot(X, B)
    return np.dot(residuals, residuals)

# Generate data
np.random.seed(42)
X = np.random.rand(50, 2)
true_beta = np.array([2, 3])
y = np.dot(X, true_beta) + np.random.normal(scale=0.1, size=50)

# Create a meshgrid for Beta1 and Beta2
beta1_values = np.linspace(1, 3, 100)
beta2_values = np.linspace(2, 4, 100)
B1, B2 = np.meshgrid(beta1_values, beta2_values)
B_mesh = np.c_[B1.ravel(), B2.ravel()]

# Compute LS values for each combination of Beta1 and Beta2
LS_values = np.array([least_squares_criterion(B, X, y) for B in B_mesh])

# Reshape LS values for contour plot
LS_values = LS_values.reshape(B1.shape)

# Create a contour plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(B1, B2, LS_values, cmap='viridis', alpha=0.8)
ax.set_xlabel('Beta1')
ax.set_ylabel('Beta2')
ax.set_zlabel('LS Value')
ax.set_title('Contour Plot of LS(Beta)')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to compute the Least Squares criterion with L1 regularization term
def lasso_least_squares_criterion(B, X, y, l1_penalty):
    residuals = y - np.dot(X, B)
    ls_term = np.dot(residuals, residuals)
    l1_term = l1_penalty * np.sum(np.abs(B))
    return ls_term + l1_term

# Generate data
np.random.seed(42)
X = np.random.rand(50, 2)
true_beta = np.array([2, 3])
y = np.dot(X, true_beta) + np.random.normal(scale=0.1, size=50)

# Create a meshgrid for Beta1 and Beta2
beta1_values = np.linspace(1, 3, 100)
beta2_values = np.linspace(2, 4, 100)
B1, B2 = np.meshgrid(beta1_values, beta2_values)
B_mesh = np.c_[B1.ravel(), B2.ravel()]

# Set L1 penalty (adjust this value)
l1_penalty = 30.

# Compute LS values with L1 regularization for each combination of Beta1 and Beta2
LS_values = np.array([lasso_least_squares_criterion(B, X, y, l1_penalty) for B in B_mesh])

# Reshape LS values for contour plot
LS_values = LS_values.reshape(B1.shape)

# Create a filled contour plot with more levels
plt.figure(figsize=(10, 8))
contour_plot = plt.contour(B1, B2, LS_values, cmap='viridis', levels=50)
plt.xlabel('Beta1')
plt.ylabel('Beta2')
plt.title('Filled Contour Plot with Lasso Term')
plt.colorbar(contour_plot, label='Objective Function Value')

# Add a diamond shape for L1 regularization term
diamond_levels = np.linspace(np.min(LS_values), np.max(LS_values), 5)  # Adjust the number of levels as needed
plt.contour(B1, B2, B1 + B2, levels=diamond_levels, colors='red', linestyles='dashed')

plt.show()
```
