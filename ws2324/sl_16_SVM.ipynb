```python
import numpy as np

class HardMarginSVM:
    def __init__(self):
        self.w = None

    def fit(self, X, y, lr=0.01, epochs=500):
        n, d = X.shape
        self.w = np.zeros(d)
        for _ in range(epochs):
            for i in range(n):
                if y[i] * np.dot(X[i], self.w) < 1:
                    self.w = self.w + lr * (y[i] * X[i])
                else:
                    self.w = self.w

    def predict(self, X):
        return np.sign(np.dot(X, self.w))

#Usage:
X = np.array([[1, 1], [2, 2], [2, 0], [0, 0], [1, 0], [0, 1]])
y = np.array([1, 1, 1, -1, -1, -1])
svm = HardMarginSVM()
svm.fit(X, y)
print(svm.predict([0.5, 0.5]))
```

```python
import cvxpy as cp
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')

class SoftMarginSVM:
    def __init__(self, C=1.0):
        self.C = C
        self.alpha = None
        self.w = None
        self.b = None
        self.support_vectors = None

    def fit(self, X, y):
        n_samples, n_features = X.shape

        # Define the optimization variables
        alpha = cp.Variable(n_samples)
        w = cp.Variable(n_features)
        b = cp.Variable()

        # Define the constraints
        constraints = [alpha >= 0,
                       #alpha <= self.C,
                       cp.sum(cp.multiply(alpha, y)) == 0,
                       cp.multiply(y, X @ w + b) >= 1]

        # Define the objective function
        objective = cp.Minimize(0.5 * cp.square(cp.norm(w)) #+ self.C * cp.sum(alpha)
                                )

        # Solve the optimization problem
        problem = cp.Problem(objective, constraints)
        problem.solve()

        # Store the solution
        self.alpha = alpha.value
        self.w = w.value
        self.b = b.value

        # Identify the support vectors
        self.support_vectors = X[self.alpha > 0]

    def predict(self, X):
        return np.sign(X @ self.w + self.b)

# Create some linearly separable data
X = np.array([[1, 2], [2, 3], [3, 3], [2, 1], [3, 2]])
y = np.array([1, 1, 1, -1, -1])

# Create and train the SVM
svm = SoftMarginSVM(C=1.0)
svm.fit(X, y)

# Use the trained SVM to make predictions
print(svm.predict(np.array([[2, 2], [3, 3]])))  # Output: [1, 1]

# Visualize the data
#plt.scatter(X[:, 0], X[:, 1], c=y)
sns.scatterplot(x = X[:, 0], y= X[:, 1], hue=y)

# Visualize the decision boundary
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = np.dot(xy, svm.w) + svm.b

# Decision boundary: w^T x + b = 0
ax.contour(XX, YY, Z.reshape(XX.shape), colors='k', levels=[0], alpha=0.5, linestyles=['-'])

# Margins: w^T x + b = 1 and w^T x + b = -1
ax.contour(XX, YY, Z.reshape(XX.shape), colors='k', levels=[-1, 1], alpha=0.5, linestyles=['--', '--'])

plt.show()
```

```python
# Exercise Sheet
# Question 2: Optimization
import numpy as np

def pegasos_linear(
        y,
        X,
        nr_iter = 50000,
        theta = np.random.normal(size=X.shape[1]),
        _lambda = 1,
        alpha = 1
):
    t = 1
    n = len(y)

    while t <= nr_iter:
        f_current = np.dot(theta.reshape(-1,1).T, X.T)
        i = np.random.randint(0,len(X))

        # update
        theta = (1 - _lambda * alpha) * theta
        if y[i]*f_current[i] < 1:
            theta = theta + alpha * y[i] * X[i,]
        t += 1

    return theta

```
