```python
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# Generate random data
n = 100
X = np.random.rand(n) * 10  # Independent variable
Z = np.random.rand(n) * 5   # Omitted variable
epsilon = np.random.normal(0, 2, n)  # Error term

# True model
Y_true = 3 * X + 2 * Z + epsilon

# Observed data (with measurement error in X)
X_observed = X + np.random.normal(0, 0.5, n)

# Estimated model (without including Z)
Y_estimated = 2 * X_observed + np.random.normal(0, 2, n)

# Scatter plot to illustrate the relationship between X and Y
plt.figure(figsize=(10, 6))
plt.scatter(X_observed, Y_estimated, label='Observed Data')
plt.scatter(X, Y_true, color='red', label='True Relationship')
plt.title('Scatter Plot of Observed Data and True Relationship')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
