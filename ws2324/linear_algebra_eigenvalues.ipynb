```python
import numpy as np

# Given matrix A
A = np.array([[0, 1],
              [1, 1]])

# Given eigenvectors
v1 = np.array([2, 1 + np.sqrt(5)])
v2 = np.array([2, 1 - np.sqrt(5)])

# Form the matrix P with columns as eigenvectors
P = np.column_stack((v1, v2))

# Find the inverse of P
P_inv = np.linalg.inv(P)

# Calculate A applied to the eigenvectors to find the eigenvalues
lambda1 = np.dot(A, v1) / v1
lambda2 = np.dot(A, v2) / v2

# Form the diagonal matrix D with eigenvalues
D = np.diag([lambda1, lambda2])

# Calculate the powers of D
D_squared = np.diag([lambda1**2, lambda2**2])
D_cubed = np.diag([lambda1**3, lambda2**3])
D_fourth = np.diag([lambda1**4, lambda2**4])

# Calculate A^n in the new basis using the diagonalization formula A^n = P D^n P_inv
A_squared = np.dot(P, np.dot(D_squared, P_inv))
A_cubed = np.dot(P, np.dot(D_cubed, P_inv))
A_fourth = np.dot(P, np.dot(D_fourth, P_inv))

# Print the results
print("Matrix P (eigenvectors):")
print(P)

print("\nMatrix P^{-1}:")
print(P_inv)

print("\nMatrix D (eigenvalues):")
print(D)

print("\nMatrix D^2:")
print(D_squared)

print("\nMatrix D^3:")
print(D_cubed)

print("\nMatrix D^4:")
print(D_fourth)

print("\nMatrix A^2 in the new basis (PD^2P^-1):")
print(A_squared)

print("\nMatrix A^3 in the new basis:")
print(A_cubed)

print("\nMatrix A^4 in the new basis:")
print(A_fourth)
```

```python
import numpy as np

# Given matrix A
A = np.array([[0, 1],
              [1, 1]])

# Given eigenvectors
v1 = np.array([2, 1 + np.sqrt(5)])
v2 = np.array([2, 1 - np.sqrt(5)])

# Normalize eigenvectors
v1_normalized = v1 / np.linalg.norm(v1)
v2_normalized = v2 / np.linalg.norm(v2)

# Form the matrix P with columns as normalized eigenvectors
P = np.column_stack((v1_normalized, v2_normalized))

# Find the inverse of P
P_inv = np.linalg.inv(P)

# Calculate A^n in the new basis using the diagonalization formula A^n = P D^n P_inv
A_squared = np.dot(P, np.dot(np.diag([1, 1]), P_inv))
A_cubed = np.dot(P, np.dot(np.diag([1, 1]), P_inv))
A_fourth = np.dot(P, np.dot(np.diag([1, 1]), P_inv))

# Convert back to the standard basis
A_squared_standard_basis = np.dot(np.dot(P_inv, A_squared), P)
A_cubed_standard_basis = np.dot(np.dot(P_inv, A_cubed), P)
A_fourth_standard_basis = np.dot(np.dot(P_inv, A_fourth), P)

# Print the results
print("Matrix A:")
print(A)

print("\nMatrix A^2 in the eigen basis:")
print(A_squared)

print("\nMatrix A^3 in the eigen basis:")
print(A_cubed)

print("\nMatrix A^4 in the eigen basis:")
print(A_fourth)

# Print the results in the standard basis
print("\nMatrix A^2 in the standard basis:")
print(A_squared_standard_basis)

print("\nMatrix A^3 in the standard basis:")
print(A_cubed_standard_basis)

print("\nMatrix A^4 in the standard basis:")
print(A_fourth_standard_basis)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data with 2 features
np.random.seed(42)
data = np.random.rand(100, 2)

# Center the data (subtract the mean of each feature)
mean_centered_data = data - np.mean(data, axis=0)

# Calculate the covariance matrix
covariance_matrix = np.cov(mean_centered_data, rowvar=False)

# Calculate the eigenvalues and eigenvectors of the covariance matrix
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)

# Sort eigenvalues and corresponding eigenvectors in descending order
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]

# Choose the top k eigenvectors based on the desired number of principal components (k)
k = 1
top_eigenvectors = eigenvectors[:, :k]

# Project the original data onto the new k-dimensional space
pca_result = np.dot(mean_centered_data, top_eigenvectors)

# Plot the original data and the PCA result
plt.scatter(data[:, 0], data[:, 1], label='Original Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.scatter(pca_result[:, 0], np.zeros_like(pca_result), label='PCA Result', marker='x')
plt.title('PCA from Scratch')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data with 2 features
np.random.seed(42)
data = np.random.rand(100, 2)
data[:, 0] = data[:, 0] + np.random.normal(loc=0, scale=1.5, size=100)

# Center the data (subtract the mean of each feature)
mean_centered_data = data - np.mean(data, axis=0)

# Calculate the covariance matrix
covariance_matrix = np.cov(mean_centered_data, rowvar=False)

# Calculate the eigenvalues and eigenvectors of the covariance matrix
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)

# Sort eigenvalues and corresponding eigenvectors in descending order
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]

# Choose the top k eigenvectors based on the desired number of principal components (k)
k = 1
top_eigenvectors = eigenvectors[:, :k]

# Project the original data onto the new k-dimensional space
pca_result = np.dot(mean_centered_data, top_eigenvectors)
pc_1 = np.dot(mean_centered_data,  eigenvectors[:, :1])
pc_2 = np.dot(mean_centered_data,  eigenvectors[:, 1:2])


# Explained variance ratio
explained_variance_ratio = eigenvalues / np.sum(eigenvalues)
print(f"Explained Variance Ratio: {explained_variance_ratio}")

# Plot the original data and the PCA result
plt.scatter(data[:, 0], data[:, 1], label='Original Data')
#plt.scatter(pca_result[:, 0], np.zeros_like(pca_result), label='PCA Result', marker='x')
plt.scatter(pc_1, pc_2, label='PCA Result', marker='x')
plt.scatter(mean_centered_data[:, 0], mean_centered_data[:, 1] , label='Mean Centered Data', marker='o')

plt.title('PCA from Scratch')
plt.legend()
plt.show()
```
