```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(42)
X = np.random.rand(100, 1) * 10  # One feature
y = 2 * X.squeeze() + np.random.randn(100)  # Target variable with noise

# Initialize parameters
learning_rate = 0.1
num_iterations = 100
predictions = np.zeros_like(y)  # Initial predictions (start with zeros)

# Gradient boosting iterations
for iteration in range(num_iterations):
    # Step 1: Calculate pseudo-residuals (negative gradients of the loss)
    residuals = y - predictions
    
    # Step 2: Fit a weak learner to the pseudo-residuals
    weak_learner = np.mean(residuals)  # For simplicity, the weak learner is the mean of residuals
    
    # Step 3: Update the base model
    predictions += learning_rate * weak_learner
    
    # Print information at every 10 iterations
    if (iteration + 1) % 10 == 0:
        mse = np.mean((y - predictions) ** 2)
        print(f"Iteration {iteration + 1}/{num_iterations} - MSE: {mse:.4f}")

# Plot the original data and the final predictions
plt.scatter(X, y, label='Original data')
plt.plot(X, predictions, color='red', label='Final Predictions')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate synthetic data
np.random.seed(42)
X = np.random.rand(100, 1)  # One feature
y = 2 * np.exp(X.squeeze()**2) + (np.random.randn(100) *0.5)  # Target variable with noise

# Initialize parameters
learning_rate = 0.05
num_iterations = 100
predictions = np.zeros_like(y)  # Initial predictions (start with zeros)
predictions = np.ones(100) * np.mean(y)


# Gradient boosting iterations
for iteration in range(num_iterations):
    # Step 1: Calculate pseudo-residuals (negative gradients of the loss)
    residuals = y - predictions
    
    # Step 2: Fit a decision tree to the pseudo-residuals
    weak_learner = LinearRegression()
    weak_learner.fit(X, residuals)
    
    # Step 3: Update the base model
    predictions += learning_rate * weak_learner.predict(X)
    plt.plot(X, predictions, color='red', alpha = iteration / num_iterations)
    
    # Print information at every 10 iterations
    if (iteration + 1) % 10 == 0:
        mse = np.mean((y - predictions) ** 2)
        print(f"Iteration {iteration + 1}/{num_iterations} - MSE: {mse:.4f}")

# Plot the original data and the final predictions
plt.scatter(X, y, label='Original data')
plt.scatter(X, predictions, color='red', label='Final Predictions')
plt.xlabel('Feature')
plt.ylabel('Target')
plt.legend()
plt.show()
```

```python
 # Second advanced model

import pandas as pd
from sklearn.tree import DecisionTreeRegressor, plot_tree

def calculate_residual(y, p):
    return y - p

def log_odds(p):
    return np.log(p/(1-p))

def probs(odds):
    return np.exp(odds) / (1+np.exp(odds))

def train_gradient_boosting_adv(x, y, n_learners, learning_rate, graph=True, describe=True):
    #f_0 = np.sum(y) / len(y) # First prediction
    f_0 = .1
    predictions_p = f_0 * np.ones(len(y))
    predictions = log_odds(predictions_p)
    print(f_0)

    for i in range(n_learners):
        print(f'Iteration: {i}')
        pred_p = probs(predictions)
        r = calculate_residual(y, pred_p)
        #r_val = r.values
        r_val = r
        tree = DecisionTreeRegressor(max_leaf_nodes=2)
        model = tree.fit(x, r)

        leaf_indices = tree.apply(x)
        terminal_output = []

        for i in range(1,3):
            output_residuals = np.sum(r_val[leaf_indices == i])
            output_predictions = np.sum(pred_p[leaf_indices == i] * (1-pred_p[leaf_indices==i]))
            output = output_residuals / output_predictions
            terminal_output.append(output)

        print(terminal_output)
        if graph == True:
            fig = plt.figure(figsize=(12,4))
            plot_tree(tree, filled=True, feature_names=x.columns)

        #predictions_i = model.predict(x)
        #print(predictions_i)
        if describe == True:
            print(f'log(odds) before {predictions}')
            print(f'Adding to predictions: {[learning_rate * terminal_output[idx - 1] for idx in leaf_indices]}')
        predictions += [learning_rate * terminal_output[idx - 1] for idx in leaf_indices]
        if describe == True:
            print(f'log(odds) after {predictions}')
            print(f'scores after {probs(predictions)}')

        zero_one_loss = [1 if j >= 0.5 else 0 for j in probs(predictions)]

        print(f'True predictions: {sum(zero_one_loss == y) / len(zero_one_loss)}')
        
        
    return model, tree
    
```
