```python
# Laplace Dist

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import laplace

# Parameters for the Laplace distributions
mu1, b1 = 50, 10
mu2, b2 = 50, 15

# Generate x values
x = np.linspace(0, 100, 1000)

# Calculate the probability density functions for both distributions
pdf1 = laplace.pdf(x, loc=mu1, scale=b1)
pdf2 = laplace.pdf(x, loc=mu2, scale=b2)

# Plot the distributions
plt.plot(x, pdf1, label=f'Laplace Distribution w/ s={b1}')
plt.plot(x, pdf2, label=f'Laplace Distribution w/ s={b2}')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Laplace Distributions with Different Scale Parameters')
plt.legend()
plt.show()
```

```python
# Laplace Dist

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import laplace, norm
from scipy.integrate import quad


# Parameters for the Laplace distributions
mu1, b1 = 0, 1.5
mu2, b2 = 0, 1

# Generate x values
x = np.linspace(-4, 4, 100)

# Calculate the probability density functions for both distributions
pdf2 = laplace.pdf(x, loc=mu1, scale=b1)
pdf1 = norm.pdf(x, loc=mu2, scale=b2)

# Define the Laplace pdf functions
pdf1_exp = lambda x: norm.pdf(x, loc=mu2, scale=b2)
pdf2_exp = lambda x: laplace.pdf(x, loc=mu1, scale=b1)

# Define the function to integrate
f_to_integrate = lambda x: pdf1_exp(x) * np.log(np.nan_to_num(pdf1_exp(x) / pdf2_exp(x)))

# Perform the numerical integration over the support [0, 100]
result, _ = quad(f_to_integrate, a=-4, b=4)

# Generate x values for visualization
integrand_values = f_to_integrate(x)


# Plot the distributions
fig, axs = plt.subplots(4,1, figsize = (6,6), sharex=True)
axs[0].plot(x, pdf2, label=f'Laplace Distribution w/ s={b1}')
axs[0].plot(x, pdf1, label=f'Normal Distribution w/ s={b2}')

axs[1].plot(x, pdf1 / pdf2, label=f'Ratio of Densities')
axs[1].set_title('Ratio of Densities')

axs[2].plot(x, np.log(pdf1 / pdf2), label=f'Ratio of Densities (log)')
axs[2].set_title('Ratio of Densities (Log)')

axs[3].plot(x, integrand_values, label=f'Integrands')
axs[3].set_title('Integrand Values')

axs[3].fill_between(x, integrand_values, color='skyblue', alpha=0.4, label='Area under the curve')


axs[0].set_ylabel('Probability Density')
axs[0].set_title('Laplace Distributions with Different Scale Parameters')
for ax in axs:
    ax.legend(loc='upper right')
    ax.set_xlabel('x')

plt.tight_layout()
plt.show()
```

```python
# Laplace Dist

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import laplace, norm
from scipy.integrate import quad


# Parameters for the Laplace distributions
mu1, b1 = 0, 3
mu2, b2 = 0, 3

# Generate x values
x = np.linspace(-4, 4, 100)

# Calculate the probability density functions for both distributions
pdf2 = laplace.pdf(x, loc=mu1, scale=b1)
pdf1 = norm.pdf(x, loc=mu2, scale=b2)

# Define the Laplace pdf functions
pdf1_exp = lambda x: norm.pdf(x, loc=mu2, scale=b2)
pdf2_exp = lambda x: laplace.pdf(x, loc=mu1, scale=b1)

# log difference
pdf1_exp_log = np.log(pdf1_exp(x))
pdf2_exp_log = np.log(pdf2_exp(x))

# exp difference
# Calculate the expectations for each PDF
expectation_pdf1 = pdf1_exp(x) * np.log(pdf1_exp(x))
expectation_pdf2 = pdf1_exp(x) * np.log(pdf2_exp(x))

exp_ratio = pdf1_exp(x) * np.log(pdf1_exp(x) / pdf2_exp(x))

# Plot the distributions
fig, axs = plt.subplots(4,1, figsize = (6,6), sharex=True)
axs[0].plot(x, pdf2, label=f'Laplace Distribution w/ s={b1}')
axs[0].plot(x, pdf1, label=f'Normal Distribution w/ s={b2}')

axs[1].plot(x, pdf1_exp_log, label=f'Log Normal')
axs[1].plot(x, pdf2_exp_log, label=f'Log Laplace')
axs[1].fill_between(x, pdf1_exp_log, pdf2_exp_log, color='skyblue', alpha=0.4, label='Area between curves')

axs[1].set_title('log[pdf1] - log(pdf2)')

axs[2].plot(x, expectation_pdf1, label=f'E[pdf1] wrt pdf1')
axs[2].plot(x, expectation_pdf2, label=f'E[pdf2] wrt pdf1')
axs[2].fill_between(x, expectation_pdf1, expectation_pdf2, color='skyblue', alpha=0.4, label='Area between curves')

axs[2].set_title('E[log[pdf1]] - E[log(pdf2)]')

axs[3].plot(x, exp_ratio, label=f'Integrands')
axs[3].set_title('E[log(pdf1/pdf2)]')

axs[3].fill_between(x, exp_ratio, color='skyblue', alpha=0.4, label='Area under the curve')


axs[0].set_ylabel('Probability Density')
axs[0].set_title('PDFs')
for ax in axs:
    ax.legend(loc='upper right')
    ax.set_xlabel('x')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, laplace
import seaborn as sns 
sns.set_style('darkgrid')

# Function to calculate KL divergence
def kl_divergence(p, q):
    return np.sum(p * np.log(p / q))

# Function to calculate negative cross entropy
def negative_cross_entropy(p, q):
    return -np.sum(p * np.log(q))

# True distribution (N(0, 1))
true_distribution = norm(loc=0, scale=1)

# Range of sigma values
sigma_values = np.arange(0.1, 5, 0.1)

# Initialize arrays to store KL and negative cross entropy values
kl_values = np.zeros_like(sigma_values)
nce_values = np.zeros_like(sigma_values)

# Calculate KL and negative cross entropy for each sigma
for i, sigma in enumerate(sigma_values):
    # Proposed distribution (Laplace(0, sigma))
    proposed_distribution = laplace(loc=0, scale=sigma)
    
    # Generate random samples for calculations
    x = np.linspace(-5, 5, 1000)
    true_pdf = true_distribution.pdf(x)
    proposed_pdf = proposed_distribution.pdf(x)
    
    # Calculate KL divergence
    kl_values[i] = kl_divergence(true_pdf, proposed_pdf)
    
    # Calculate negative cross entropy
    nce_values[i] = negative_cross_entropy(true_pdf, proposed_pdf)

# Plot KL divergence
fig, axs = plt.subplots(1,2, figsize=(10, 5), sharex=True, sharey=True)
axs[0].plot(sigma_values, kl_values, color='blue', label='KL Divergence')
axs[0].set_xlabel('Sigma')
axs[0].set_ylabel('KL Divergence')
axs[0].set_title('KL Divergence between N(0, 1) and Laplace(0, sigma)')

# Plot negative cross entropy
axs[1].plot(sigma_values, nce_values, color='red', label='Negative Cross Entropy')
axs[1].plot(sigma_values, kl_values, color='blue', label='KL Divergence')

axs[1].set_xlabel('Sigma')
axs[1].set_ylabel('Negative Cross Entropy')
axs[1].set_title('Negative Cross Entropy between N(0, 1) and Laplace(0, sigma)')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.feature_selection import mutual_info_classif

# Generate a synthetic dataset
X, y = make_classification(n_samples=1000, n_features=20, n_informative=10, n_redundant=5, random_state=42)

# Calculate mutual information between each feature and the target variable
mi_scores = mutual_info_classif(X, y, discrete_features=False, random_state=42)

# Plot the mutual information scores for each feature
plt.figure(figsize=(10, 6))
plt.bar(range(X.shape[1]), mi_scores, align='center', alpha=0.7)
plt.xticks(range(X.shape[1]), [f'Feature {i+1}' for i in range(X.shape[1])])
plt.xlabel('Features')
plt.ylabel('Mutual Information Score')
plt.title('Mutual Information between Features and Target Variable')
plt.show()
```
