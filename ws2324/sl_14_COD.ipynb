```python
import matplotlib.pyplot as plt
import numpy as np

# Define the range of fractions (from 0.01 to 1 for better visibility)
fractions = np.linspace(0.01, 1, 100)

# Calculate the corresponding edge lengths for 1, 2, and 3 dimensions
edge_length_1d = 1 / fractions
edge_length_2d = 1 / np.sqrt(fractions)
edge_length_3d = 1 / np.cbrt(fractions)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(fractions, edge_length_1d, label='1 Dimension')
plt.plot(fractions, edge_length_2d, label='2 Dimensions')
plt.plot(fractions, edge_length_3d, label='3 Dimensions')

# Add labels and title
plt.xlabel('Fraction of Volume Covered')
plt.ylabel('Edge Length of Smaller Object')
plt.title('Relationship between Fraction of Volume Covered and Edge Length')
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_style('darkgrid')

# Define the range of fractions (from 0.01 to 1 for better visibility)
fractions = np.linspace(0.01, 1, 100)

# Calculate the corresponding edge lengths for 1, 2, and 3 dimensions
edge_length_1d = fractions
edge_length_2d = np.sqrt(fractions)
edge_length_3d = np.cbrt(fractions)
edge_length_10d = np.power(fractions, 1/10)


# Plotting
plt.figure(figsize=(10, 6))
plt.plot(fractions, edge_length_1d, label='1 Dimension')
plt.plot(fractions, edge_length_2d, label='2 Dimensions')
plt.plot(fractions, edge_length_3d, label='3 Dimensions')
plt.plot(fractions, edge_length_10d, label='10 Dimensions')


# Add labels and title
plt.xlabel('Fraction of Volume Covered')
plt.ylabel('Edge Length of Smaller Object')
plt.title('Relationship between Fraction of Volume Covered and Edge Length')
plt.legend()
plt.axvline(0.10, color='grey', linestyle='dashed')
# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the range of edge lengths (from 0.01 to 1 for better visibility)
edge_lengths = np.linspace(0.01, 1, 100)

# Define different dimensions for hyperspheres
dimensions = [1, 2, 3, 10]

# Plotting
plt.figure(figsize=(10, 6))

for p in dimensions:
    covered_volumes = 0.5**p - (edge_lengths / 2)**p
    plt.plot(edge_lengths, covered_volumes, label=f'Hypersphere in Hypercube (p={p})')

# Add labels and title
plt.xlabel('Edge Length of Hypercube')
plt.ylabel('Fraction of Volume Covered')
plt.title('Relationship between Edge Length and Covered Volume for Hyperspheres in Hypercubes')
plt.legend()

# Show the plot
plt.show()
```
